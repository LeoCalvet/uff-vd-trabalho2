## Primeiro grfico

>**Motivao para a Escolha da Visualizao:** A visualizao foi escolhida para analisar a diferena de avaliaes (positivas e negativas) dos jogos na plataforma Steam. A diferena de avaliaes  uma mtrica importante que pode indicar a popularidade e a aceitao de um jogo pelos usurios. Alm disso, a visualizao tambm permite analisar essa mtrica em relao aos diferentes gneros de jogos. Isso pode fornecer insights sobre quais gneros de jogos so mais bem recebidos pelos usurios.

>**Anlise das Informaes da Visualizao:** A visualizao fornece uma srie de informaes teis. Para cada gnero de jogo, ela mostra os cinco principais jogos com a maior diferena de avaliaes. Isso pode ajudar a identificar quais jogos so os mais populares em cada gnero. Alm disso, a diferena absoluta de avaliaes (avaliaes positivas menos avaliaes negativas)  usada como uma mtrica de sucesso do jogo. Isso fornece uma medida mais equilibrada do sucesso, pois leva em considerao tanto as avaliaes positivas quanto as negativas. Por fim, a visualizao usa um grfico de barras, que  uma escolha eficaz para comparar quantidades entre diferentes categorias. Neste caso, ele permite uma comparao clara da diferena de avaliaes entre os cinco principais jogos de cada gnero.

```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets

file_path = '/content/steam.csv'
steam_data = pd.read_csv(file_path)
steam_data['rating_difference'] = abs(steam_data['positive_ratings'] - steam_data['negative_ratings'])
genres_frequency = steam_data['genres'].str.split(';', expand=True).stack().value_counts()
top_10_genres = genres_frequency.head(10).index.tolist()
top_games_by_genres = {}
for genres in top_10_genres:
    genres_games = steam_data[steam_data['genres'].str.contains(genres)]
    top_games = genres_games.sort_values(by='rating_difference', ascending=False).head(5)
    top_games_by_genres[genres] = top_games[['name', 'rating_difference']]

def plot_genres(genres):
    games = top_games_by_genres[genres]
    plt.figure(figsize=(10, 6))
    plt.bar(games['name'], games['rating_difference'], color='skyblue')
    plt.xlabel('Nome do Jogo')
    plt.ylabel('Diferena de Avaliao (Absoluta)')
    plt.title(f'Top 5 Jogos Mais Bem Avaliados no Gnero {genres}')
    plt.xticks(rotation=45)
    plt.show()


genres_dropdown = widgets.Dropdown(options=top_10_genres)

widgets.interact(plot_genres, genres=genres_dropdown)
```

## Segundo grfico

>**Motivao para a Escolha da Visualizao:** Um grfico de disperso interativo que mostra a relao entre o preo e o tempo mdio de jogo dos 50 jogos mais bem avaliados na plataforma Steam. A escolha dessa visualizao foi motivada pela necessidade de entender como essas duas variveis se relacionam entre os jogos mais populares. Os grficos de disperso so uma excelente ferramenta para visualizar a relao entre duas variveis numricas. Eles permitem identificar correlaes, padres e outliers nos dados. Alm disso, a adio de interatividade (exibir o nome do jogo ao passar o mouse sobre um ponto) melhora a utilidade do grfico, permitindo que os usurios obtenham mais detalhes sobre pontos especficos de interesse.

>**Anlise das Informaes da Visualizao:** A visualizao fornece vrias informaes teis. Em primeiro lugar, ela mostra a distribuio dos preos e do tempo mdio de jogo entre os 50 jogos mais bem avaliados. Isso pode ajudar a identificar se os jogos mais caros tendem a ter tempos de jogo mais longos, ou se no h uma correlao clara entre essas duas variveis. Alm disso, a visualizao permite identificar quais jogos oferecem o melhor valor em termos de tempo de jogo pelo preo. Por exemplo, os jogos que esto mais  direita do grfico (indicando um tempo de jogo mais longo) mas que no so muito caros podem ser considerados como tendo um bom valor. Finalmente, a visualizao permite identificar quaisquer outliers nos dados. Por exemplo, se houver algum jogo que seja muito mais caro ou que tenha um tempo de jogo muito mais longo do que os outros jogos, isso ser claramente visvel no grfico.

```python
import pandas as pd
import altair as alt

steam_data = pd.read_csv('/content/steam.csv')

steam_data['rating_difference'] = steam_data['positive_ratings'] - steam_data['negative_ratings']

top_50_games = steam_data.nlargest(50, 'rating_difference')

scatter_plot = alt.Chart(top_50_games).mark_circle(size=60).encode(
    x=alt.X('price:Q', title='Preo'),
    y=alt.Y('average_playtime:Q', title='Tempo Mdio de Jogo'),
    tooltip=['name:N', 'price:Q', 'average_playtime:Q']
).properties(
    width=700,
    height=400,
    title='Top 50 Jogos Mais Bem Avaliados: Preo vs Tempo Mdio de Jogo'
).interactive()

scatter_plot
```

## Terceiro grfico

>**Motivao para a Escolha da Visualizao:** Um grfico de rea empilhado interativo que mostra a quantidade de jogos lanados ao longo do tempo para os 5 gneros mais populares na plataforma Steam. A escolha dessa visualizao foi motivada pela necessidade de entender como a popularidade dos gneros de jogos mudou ao longo do tempo. Os grficos de rea empilhados so uma excelente ferramenta para visualizar a distribuio de categorias ao longo do tempo. Eles permitem ver no apenas a mudana total ao longo do tempo, mas tambm como a composio dessa totalidade mudou. Alm disso, a adio de interatividade (exibir o gnero ao passar o mouse sobre uma rea) melhora a utilidade do grfico, permitindo que os usurios obtenham mais detalhes sobre reas especficas de interesse.

>**Anlise das Informaes da Visualizao:** A visualizao fornece vrias informaes teis. Em primeiro lugar, ela mostra a distribuio dos gneros de jogos ao longo do tempo. Isso pode ajudar a identificar tendncias, como se certos gneros esto se tornando mais populares ou menos populares. Alm disso, a visualizao permite ver a proporo de jogos de cada gnero em qualquer ano. Isso pode fornecer insights sobre a diversidade de gneros em diferentes perodos de tempo. Finalmente, a visualizao permite identificar quaisquer picos ou quedas na quantidade total de jogos lanados. Por exemplo, se houver um pico em um determinado ano, isso pode indicar um ano particularmente bom para o lanamento de jogos.

```python
import pandas as pd
import altair as alt

# Carregar os dados
steam_data = pd.read_csv('/content/steam.csv')

# Converter a data de lanamento para o formato datetime e extrair o ano
steam_data['release_date'] = pd.to_datetime(steam_data['release_date'])
steam_data['year'] = steam_data['release_date'].dt.year

# Obter os 5 gneros mais populares
top_5_genres = steam_data['genres'].value_counts().nlargest(5).index.tolist()

# Filtrar os dados para incluir apenas os jogos dos 5 gneros mais populares
filtered_data = steam_data[steam_data['genres'].isin(top_5_genres)]

# Agrupar os dados por gnero e ano e contar o nmero de jogos
grouped_data = filtered_data.groupby(['genres', 'year']).size().reset_index(name='count')

# Criar um grfico de rea empilhado
alt.Chart(grouped_data).mark_area().encode(
    x="year:T",
    y=alt.Y("count:Q", stack=True),
    color="genres:N",
    tooltip=['genres', 'year', 'count']
).interactive()
```
